```python
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

# Perulangan for untuk memeriksa setiap nilai vibrasi
for nomor_unit, nilai_vibrasi in enumerate(vibrasi_mesin, start=1):
    if nilai_vibrasi > 2.0:
        print(f"Warning: Unit {nomor_unit} vibrasi tinggi!")
```

```python
# Meminta input suhu saat ini dari pengguna
suhu_saat_ini = float(input("Masukkan suhu saat ini (°C): "))

# Loop while untuk menaikkan suhu hingga mencapai target
while suhu_saat_ini < 85:
    suhu_saat_ini += 5  # Menaikkan suhu 5 derajat setiap iterasi
    print(f"Suhu saat ini: {suhu_saat_ini}°C")

print("Target suhu 85°C telah tercapai!")
```

```python
# Meminta input dari pengguna
kapasitas_maksimum = float(input("Masukkan kapasitas maksimum (kW): "))
beban_terpakai = float(input("Masukkan beban terpakai (kW): "))

# Menghitung persentase pemakaian
persentase = (beban_terpakai / kapasitas_maksimum) * 100

# Mengklasifikasikan status beban berdasarkan persentase
if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"

# Menampilkan hasil
print(f"\nKapasitas maksimum: {kapasitas_maksimum} kW")
print(f"Beban terpakai: {beban_terpakai} kW")
print(f"Persentase pemakaian: {persentase:.2f}%")
print(f"Status: {status}")
```

```python
# Data suhu boiler
suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

# Loop for untuk menampilkan setiap nilai suhu
print("Monitoring Suhu Boiler:")
print("-" * 25)

for i, suhu in enumerate(suhu_boiler, start=1):
    print(f"Titik {i}: {suhu}°C", end="")

    # Memeriksa apakah suhu di atas 100°C
    if suhu > 100:
        print(" ⚠️  PERINGATAN: Suhu di atas 100°C!")
    else:
        print()  # Pindah baris untuk suhu normal
```
