```python
def hitung_efisiensi(input_energi, output_energi):
    """
    Menghitung efisiensi turbin berdasarkan input dan output energi.

    Parameters:
    input_energi (float): Energi input ke sistem
    output_energi (float): Energi output dari sistem

    Returns:
    float: Nilai efisiensi (output / input)
    """
    efisiensi = output_energi / input_energi
    return efisiensi

# Contoh pemanggilan fungsi dengan input 500 dan output 400
hasil_efisiensi = hitung_efisiensi(500, 400)
print(f"Efisiensi turbin: {hasil_efisiensi:.2f} atau {hasil_efisiensi*100:.1f}%")
```

```python
def hitung_efisiensi_termal(energi_masuk, energi_keluar):
    """
    Menghitung efisiensi termal turbin gas.

    Parameters:
    energi_masuk (float): Energi yang masuk ke sistem (dalam Joule)
    energi_keluar (float): Energi yang keluar dari sistem (dalam Joule)

    Returns:
    float: Efisiensi termal dalam persen (%)
    """
    if energi_masuk <= 0:
        return 0  # Menghindari pembagian dengan nol atau nilai negatif

    efisiensi = (energi_keluar / energi_masuk) * 100
    return efisiensi

# Contoh penggunaan fungsi
energi_input = float(input("Masukkan energi masuk (Joule): "))
energi_output = float(input("Masukkan energi keluar (Joule): "))

efisiensi = hitung_efisiensi_termal(energi_input, energi_output)

print(f"Efisiensi termal turbin gas: {efisiensi:.2f}%")
```

```python
# Simulasi pengisian tangki bahan bakar
level_tangki = 0  # Mulai dari 0 liter

while level_tangki < 100:
    level_tangki += 10  # Naikkan 10 liter setiap iterasi

    # Memeriksa kondisi level tangki
    if level_tangki == 80:
        print("WARNING: Hampir penuh!")
    elif level_tangki == 100:
        print("TANGKI PENUH - Pengisian berhenti.")
    else:
        print(f"Level tangki: {level_tangki} liter")
```
